```python
from efficient_apriori import apriori
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
from PyARMViz import Rule
from PyARMViz import datasets
from PyARMViz import PyARMViz

import warnings
warnings.filterwarnings('ignore')

plt.rcParams['font.sans-serif'] = ['SimHei']  # 用来正常显示中文标签
plt.rcParams['axes.unicode_minus'] = False  # 用来正常显示负号

data1=pd.read_csv("2020/survey_results_public.csv")
plat=data1["PlatformWorkedWith"]
db=data1["DatabaseWorkedWith"]
frame=data1["WebframeWorkedWith"]
lang=data1["LanguageWorkedWith"]
misc=data1["MiscTechWorkedWith"]
newc=data1["NEWCollabToolsWorkedWith"]
```

```python
items=[plat,db,frame,lang,misc,newc]
trans=[]

def process(x):
    l=[]
    for i in x:
        ll=i.split(";")
        l.append(ll)
    return l
for i in items:
    i.dropna(inplace=True)   
    trans.append(process(i))
trans
```

```python
def cov(list,min_support,min_confidence,min_lift):
    itemsets, rules = apriori(list, min_support=min_support,min_confidence=min_confidence)
    print("--------------------------------")
    print(itemsets)
    print("--------------------------------")
    print(rules)
    print("--------------------------------")
    result = sorted(rules, key=lambda rule: rule.lift,reverse=True)
    support_list=[]
    confidence_list=[]
    lift_list=[]
    for i in range(len(result)):
        rule=result[i]
        if rule.lift>min_lift:
            print(rule)
            support_list.append(rule.support)
            confidence_list.append(rule.confidence)
            lift_list.append(rule.lift)
        else:
            result=result[:i]
            break
    PyARMViz.metadata_scatter_plot(result,True)
    PyARMViz.adjacency_scatter_plot(result)
    PyARMViz.adjacency_parallel_coordinate_plot(result)
    
print("platform:")
cov(trans[0],0.15,0.25,1)
plt.show()
```
