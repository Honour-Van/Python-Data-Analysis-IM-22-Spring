```python
# 清理性别部分的数据，将同时选择了labels列表中的数据合并为第三性别“Non-binary, genderqueer, or gender non-conforming”
idx = s2020[s2020['Gender']=='Woman;Man;Non-binary, genderqueer, or gender non-conforming'].index
s2020.drop(index=idx,inplace=True)
sex = s2020['Gender'].value_counts()
labels = ['Man;Non-binary, genderqueer, or gender non-conforming','Woman;Non-binary, genderqueer, or gender non-conforming','Woman;Man']
for label in labels:
    sex['Non-binary, genderqueer, or gender non-conforming'] += sex[label]
    sex.drop(label,inplace=True)
```

```python
'''
整理种族部分的数据，由于问卷此题是多选，因此使用统计出现的种族名称次数的办法，即如果选了2个种族，归类为'Biracial'，3个及以上归类为'Multiracial'
'''
s2020.dropna(subset=['Ethnicity'],inplace=True)
eth = pd.DataFrame(columns=['Ethnicity','Counts'])
index=['Black or of African descent','East Asian','Hispanic or Latino/a/x','Middle Eastern',
                            'White or of European descent','Biracial','Indigenous (such as Native American, Pacific Islander, or Indigenous Australian)','South Asian','Multiracial',
                            'Southeast Asian','I don\'t know','Prefer not to say']
eth['Ethnicity'] = index
dic = {}
cnt = [0 for i in range(12)]
j=0
for i in index:
    dic[i] = j
    j += 1
print(dic)
#s2020.loc[53,'Ethnicity'].split(';')

for i in range(len(s2020)):
    try:
        eth_tmp = s2020.iloc[i,16].split(';')
        n = len(eth_tmp)
        if n == 1:
            idx = dic[eth_tmp[0]]
            cnt[idx] += 1
        elif n==2:
            cnt[5] += 1
        else:
            cnt[8] +=1
    except:
        eth_tmp = s2020.iloc[i,16]
        n = len(eth_tmp)
        if n == 1:
            idx = dic[eth_tmp[0]]
            cnt[idx] += 1
        elif n==2:
            cnt[5] += 1
        else:
            cnt[8] +=1

eth['Counts'] = cnt
```

### 5.6 受访者选择职业时的考虑及更换工作时的考虑

```python
aspects = ['Remote work options',
           'Office environment or company culture',
           'Financial performance or funding status of the company or organization',
           'Opportunities for professional development',
          'Diversity of the company or organization',
           'How widely used or impactful my work output would be',
          'Industry that I’d be working in',
          'Specific department or team I’d be working on',
          'Flex time or a flexible schedule',
           'Languages, frameworks, and other technologies I’d be working with',
          'Family friendliness']
factor = s2020['JobFactors'].str.split(';', expand=True)
cnt = factor.stack().value_counts(ascending=True)
plt.barh(cnt.index,cnt,alpha=0.7,color='#957DAD')
```

```python
s2020.reset_index(drop=True,inplace=True)
def split_multicolumn(col_series):
    result_df = col_series.to_frame()
    options = []
    # Iterate over the column
    for idx, value  in col_series[col_series.notnull()].iteritems():
        # Break each value into list of options
        for option in value.split(';'):
            # Add the option as a column to result
            if not option in result_df.columns:
                options.append(option)
                result_df[option] = False
            # Mark the value in the option column as True
            result_df.at[idx, option] = True
    return result_df[options]
```

```python
mapping = {True:1,False:0}
def find_aspect_ratio(df):
    dic = {}
    for aspect in aspects:
        tmp = df[aspect]
        dic[aspect] = np.mean(tmp.map(mapping))
    aspect_df = pd.DataFrame(list(dic.items()),columns=['aspect','proportion']).set_index(['aspect']).sort_values('proportion',ascending=False)     
    return aspect_df
```

```python
s2020['YearsCodePro'] = pd.to_numeric(s2020['YearsCodePro'], errors='coerce').fillna(0)
```

```python
def coding_experience(s):
    if s>=10:
        return 1
    else:
        return 0
s2020['CodingExperience'] = s2020["YearsCodePro"].apply(coding_experience)

df_high_coding_experience = s2020.loc[s2020['CodingExperience'] == 1]
df_low_coding_experience = s2020.loc[s2020['CodingExperience'] == 0]
```

```python
def table_plot(df1, df2, name1, name2):
    """
    拼接并输出表格
    """

    ratio_1_perc = find_aspect_ratio(split_multicolumn(df1.JobFactors))
    ratio_2_perc = find_aspect_ratio(split_multicolumn(df2.JobFactors))

    comp_df = pd.merge(ratio_1_perc, ratio_2_perc, left_index=True, right_index=True)
    comp_df.columns = [name1, name2]
    comp_df['Fractional difference'] = (comp_df[name1] - comp_df[name2])/(comp_df[name1])
    out = comp_df.style.bar(subset=['Fractional difference'], align='mid', color=['#D6DAF9', '#FFDEEB'])

    return out
```

```python
'''
表格意为将受访者分为专业编写代码年限超过10年的和没超过10年的，观察哪个工作因素对受访者寻找工作时影响较大，aspect列从高到低排列，可以看出，
绝大多数受访者在寻找工作时都会考虑工作时选择的语言和技术，其次是灵活的工作时间，并且工作经验超过10年的受访者会更关注灵活工作时间这一因素

'''
table_plot(df_high_coding_experience,df_low_coding_experience,'> 10 yr working experience', '< 10 yr working experience')
```

```python

def create_color_dic(palette,labels):
    color_dic={}
    i = 0
    p=0
    n = len(labels)//len(palette)
    n += 1
    for label in labels:
        color_dic[label] = palette[p]
        i += 1
        if i == n:
            i = 0
            p += 1
    return color_dic
```
